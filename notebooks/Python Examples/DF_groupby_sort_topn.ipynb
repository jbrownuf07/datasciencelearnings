```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'State': [
        'California',
        'California',
        'California',
        'California',
        'Kentucky',
        'Kentucky',
        'Kentucky',
        'New York',
        'New York'
    ],
    'Sector': [
        1, 2, 3, 2, 2, 1, 3, 1, 3
    ],
    'Name': [
        'Tom',
        'Harry',
        'Roger',
        'Jim', 
        'Bob',
        'Roger',
        'Jill',
        'Sally',
        'Harry'
    ],
    'Revenue': [
        10,
        20,
        30,
        40,
        15,
        25,
        45,
        50,
        15
    ]
})

df
```

```python
df.groupby('State') \
    .apply(lambda grp: grp.nlargest(2, 'Revenue')) \
    .droplevel(level=1, axis=0)
```

```python
df.groupby('State').apply(lambda grp: grp.nlargest(2, 'Sector'))
```
