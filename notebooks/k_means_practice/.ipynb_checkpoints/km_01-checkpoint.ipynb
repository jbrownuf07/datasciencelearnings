# Practicing K-means clustering

```python
%matplotlib inline

from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
DATA_FILE_PATH = '../../data/xclara.csv'
```

```python
df = pd.read_csv(DATA_FILE_PATH)
print(df.shape)
df.head()
```

```python
# Get the data values and plot the data
f1 = df['V1'].values
f2 = df['V2'].values
X = np.array(list(zip(f1, f2)))

plt.scatter(f1, f2, c='black', s=7)
```

```python
# Randomly assign centroids
np.random.seed(93)

# Number of cluster
k = 7

c_dimensions = []

for i in range(k):
    c_dimensions.append(np.random.randint(X.max() - 20, size = df.shape[1]))
C = np.array(np.vstack(c_dimensions))

C
```

```python
# plot data with the starting centroids
plt.scatter(f1, f2, color = 'black', s = 7)
plt.scatter(C[:,0], C[:,1], color = 'red', s = 200, marker = '*')
```

```python
def dist(a, b, ax=1):
    return np.linalg.norm(a - b, axis = ax)
```

```python
# Intialize array to store old centroids
C_old = np.zeros(C.shape)

# Store cluster labels
clusters = np.zeros(len(X))

error = dist(C, C_old, None)

error
```

```python
while error != 0:
    
    for i in range(len(X)):
        distances = dist(X[i], C)
        cluster = np.argmin(distances)
        clusters[i] = cluster
        
    C_old = deepcopy(C)
    
    for i in range(k):
        points = [X[j] for j in range(len(X)) if clusters[j] == i]
        C[i] = np.mean(points, axis = 0)
    
    error = dist(C, C_old, None)
```

```python
C
```

```python
colors = ['r', 'g', 'b', 'c', 'y', 'm', 'orange']

fig, ax = plt.subplots()
for i in range(k):
    points = np.array([X[j] for j in range(len(X)) if clusters[j] == i])
    ax.scatter(points[:,0], points[:,1], color = colors[i], s = 7)
    
ax.scatter(C[:,0], C[:,1], color = 'black', s=200, marker = '*')
```
